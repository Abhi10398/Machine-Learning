```
import numpy as np
import matplotlib.pyplot as plt

class regression:
  # Constructor
  def __init__(self, name='reg'):
    self.name = name  # Create an instance variable

  def grad_update(self,w_old,lr,y,x):
    # w=w_old+2*lr*(x@(y-x.T@w_old))/len(y)
    w=w_old+2*lr*np.dot(x,(y-np.dot(np.transpose(x),w_old)))/len(y)
    return w

  def error(self,w,y,x):
    return sum((np.dot(np.transpose(x),w)-y)**2)/len(y) 

  def mat_inv(self,y,x_aug):
    return np.dot(np.dot(np.linalg.inv(np.dot(x_aug,np.transpose(x_aug))),x_aug),y)
    # by Gradien descent
  def Regression_grad_des(self,x,y,lr):
    m,n=x.shape
    w=np.random.randn(m,1)
    err=[]
    for _ in range(1000):
      err.append(self.error(w,y,x))
      w=self.grad_update(w,lr,y,x)
    w_pred=w

    return w_pred,err
```

```
# USING GRADIENT DESCENT METHOD
def data_transform(X,degree): 
  X_new_norm=np.array([(X**i)/(10**i) for i in reversed(range(degree+1))])
  return X_new_norm

x=np.linspace(-6,6,100)
y=0.25*(x**3)+1.25*(x**2)-3*x-3
y=y+3*np.random.randn(100)
plt.plot(x,y,".")

# for degree 3 polynomial
X_1=data_transform(x,3)
Y=y/100
Y=Y.reshape(X_1.shape[1],1)
w_mat,_=reg.Regression_grad_des(X_1,Y,0.6)
y_pred=(X_1.T @ w_mat) * 100
plt.figure()
plt.plot(x.T,y,'.')
plt.plot(x.T,y_pred)
```
